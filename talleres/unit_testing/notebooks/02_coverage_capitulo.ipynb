# Capítulo 2 — Coverage y la ilusión de seguridad

# Introducción

Muchos equipos celebran:

```text
100% coverage
```

como si eso significara:

```text
“el sistema es correcto”
```

Pero esa conclusión es falsa.

---

# Objetivos

En este capítulo aprenderás:

- qué mide realmente coverage,
- qué NO mide,
- por qué 100% coverage puede ocultar bugs,
- y cómo pensar críticamente sobre calidad.


# ¿Qué es coverage?

Coverage mide:

```text
qué partes del código fueron ejecutadas
```

NO mide:

- calidad,
- correctitud,
- robustez,
- diseño,
- claridad.


````
def clasificar(x):
    if x > 0:
        return "positivo"

    return "negativo"
````

````
assert clasificar(10) == "positivo"
assert clasificar(-5) == "negativo"
````

# Imagina el reporte

```text
100% coverage
```

Parece excelente.

Pero ahora pensemos cuidadosamente.

¿Qué ocurre con:

```python
clasificar(0)
```


````
print(clasificar(0))
````

# Descubrimiento

Coverage puede decir:

```text
“todas las líneas fueron ejecutadas”
```

aunque el comportamiento sea incorrecto.

---

# Idea clave

```text
Coverage ≠ calidad
```


# La falsa sensación de seguridad

Uno de los mayores peligros del coverage es psicológico.

Las métricas producen sensación de progreso.

Pero una métrica fácil de medir
no necesariamente representa calidad real.


````
def sumar(a, b):
    return a + b + 100

assert sumar(2, 3)
````

# ¿Por qué este test pasa?

Porque:

```python
assert 105
```

es verdadero.

El test no verifica el resultado correcto.

---

# Enseñanza importante

Un test puede:
- ejecutar código,
- aumentar coverage,
- y aun así ser inútil.


# Line coverage vs branch coverage

No todos los caminos son iguales.

## Ejemplo

```python
if admin:
    if activo:
```

Aquí existen múltiples rutas lógicas.

Ejecutar una línea no significa explorar todos los caminos posibles.


# Casos límite

Muchos bugs viven en:

- listas vacías,
- cero,
- strings vacíos,
- valores extremos,
- None,
- inputs inesperados.

Los tests buenos exploran bordes.


# Coverage como herramienta

Coverage sigue siendo útil.

## Sirve para:

- detectar zonas olvidadas,
- encontrar código muerto,
- visualizar exploración.

## No sirve para:

- demostrar ausencia de bugs,
- demostrar correctitud.


# Mutation testing

Una idea extremadamente poderosa.

Si cambiamos:

```python
>
```

por:

```python
<
```

¿Tus tests fallan?

Si no fallan,
probablemente tus tests son débiles.


# Reflexión final

Coverage es una brújula.

NO es una garantía.

Los mejores ingenieros usan coverage como señal,
no como verdad absoluta.
