# Capítulo 4 — El lado oscuro de los tests

# Introducción

Existe una idea peligrosa en programación:

```text
“todos los tests pasan”
```

Muchos equipos interpretan eso como:

```text
“el sistema funciona”
```

Pero eso es falso.

Los tests pueden:

- mentir,
- ser débiles,
- ocultar bugs,
- generar falsa confianza.

---

# Idea central

El testing NO es escribir asserts.

El testing es diseñar mecanismos confiables
para detectar comportamiento incorrecto.


# Objetivos

En este capítulo aprenderás:

- qué son los falsos positivos,
- cómo los tests pueden mentir,
- por qué existen tests débiles,
- cómo pensar sobre calidad real,
- y cómo usar property-based testing.


````
def promedio(lista):
    return sum(lista) / len(lista) + 1
````

````
assert promedio([1, 2, 3])
````

# El problema

El test pasa.

Pero la implementación está mal.

---

# Pregunta importante

¿Por qué ocurre esto?

Porque el test solo verifica:

```python
assert valor
```

Es decir:

```text
“existe un resultado”
```

NO:

```text
“el resultado es correcto”
```


````
print(promedio([1,2,3]))
````

# Falsos positivos

Un falso positivo ocurre cuando:

```text
los tests pasan
PERO
el sistema está roto
```

Este es uno de los mayores peligros del testing.


# Tests débiles

Muchos tests parecen correctos
pero verifican muy poco.

## Ejemplos peligrosos

```python
assert resultado

assert lista

assert response
```

Estos tests suelen validar existencia,
no correctitud.


````
assert promedio([1,2,3]) == 2
````

# Mejorando el test

Ahora verificamos comportamiento correcto.

Pero todavía hay un problema.

---

# Pregunta

¿Qué casos faltan?

- listas vacías,
- negativos,
- floats,
- un elemento,
- números enormes,
- valores aleatorios.


# El problema de ejemplos manuales

Una de las limitaciones humanas es que:

```text
pensamos en pocos casos
```

Los sistemas reales reciben millones de inputs posibles.


# Property-based testing

Aquí aparece una idea revolucionaria.

En vez de escribir ejemplos específicos:

```python
assert promedio([1,2,3]) == 2
```

describimos propiedades generales.


````
from hypothesis import given
from hypothesis import strategies as st

@given(st.lists(st.integers(), min_size=1))
def test_promedio(lista):

    resultado = promedio(lista)

    assert min(lista) <= resultado <= max(lista)
````

# ¿Qué hace Hypothesis?

Hypothesis genera automáticamente:

- listas,
- enteros,
- casos raros,
- bordes,
- inputs inesperados.

# Pregunta

¿Tus tests detectarían el error?

Si NO lo detectan,
tus tests probablemente son débiles.


# Anti-patrones comunes

## Tests demasiado específicos

Dependen de implementación interna.

## Tests frágiles

Fallan por razones irrelevantes.

## Tests lentos

Nadie quiere ejecutarlos.

## Tests acoplados

Un test rompe otros tests.


# Calidad real

La calidad NO se mide solamente por:

- cantidad de tests,
- porcentaje de coverage,
- cantidad de asserts.

La verdadera calidad depende de:

- capacidad de detectar bugs,
- claridad,
- mantenibilidad,
- confianza.


# Reflexión final

Un buen test:

- falla cuando debe fallar,
- detecta cambios peligrosos,
- explica comportamiento,
- ayuda a entender el sistema.

---

# Idea final

El objetivo del testing no es:

```text
hacer que todo pase
```

El objetivo es:

```text
detectar errores reales antes de producción
```
