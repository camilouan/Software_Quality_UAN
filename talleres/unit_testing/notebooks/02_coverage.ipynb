# Cobertura de Código

Cobertura ≠ calidad.

```
def clasificar(x):
    if x > 0:
        return 'positivo'
    else:
        return 'no positivo'
```

```
def test_clasificar():
    assert clasificar(5) == 'positivo'
```

## Pregunta
¿Qué falta probar?

```
def test_negativo():
    assert clasificar(-1) == 'no positivo'
```
