# Introducción a Mocks

## ¿Qué es un mock?
Permite simular dependencias externas (APIs, archivos, BD).

```python
import requests

def obtener_texto(url):
    return requests.get(url).text
```

## Problema
¿Qué pasa si no hay internet?

```python
def test_obtener_texto():
    texto = obtener_texto('https://example.com')
    assert 'Example' in texto
```

## Uso de mocks

```python
from unittest.mock import patch, Mock

def test_mock():
    fake = Mock()
    fake.text = 'hola mundo'
    with patch('requests.get', return_value=fake):
        assert obtener_texto('fake') == 'hola mundo'
```

## Stub vs Mock
- Stub: devuelve datos
- Mock: verifica comportamiento
